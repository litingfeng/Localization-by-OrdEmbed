```python
import matplotlib.pyplot as plt
import numpy as np
from anchors import generate_anchors
import torch

a = generate_anchors(base_size=16, ratios=[0.5, 1.0, 2.0, 3.0, 4.0],
                             scales=np.array([1.5, 3.0, 6.0, 9.0, 12.0]))
```

```python
 #print(a)
feat_height, feat_width = 14, 14 # 81/4
shift_x = np.arange(0, feat_width) * 16
shift_y = np.arange(0, feat_height) * 16
shift_x, shift_y = np.meshgrid(shift_x, shift_y)
shifts = torch.from_numpy(np.vstack((shift_x.ravel(), shift_y.ravel(),
                          shift_x.ravel(), shift_y.ravel())).transpose())
shifts = shifts.contiguous().float()
print('shifts ', shifts)

A = a.shape[0]
K = shifts.size(0)

_anchors = torch.from_numpy(a)
# anchors = self._anchors.view(1, A, 4) + shifts.view(1, K, 4).permute(1, 0, 2).contiguous()
anchors = _anchors.view(1, A, 4) + shifts.view(K, 1, 4)
print('anchors ', anchors.shape)
anchors = anchors.view(K * A, 4)
anchors = torch.clip(anchors, 0, 223)
# filter small boxes
ws = anchors[:, 2] - anchors[:, 0] + 1
hs = anchors[:, 3] - anchors[:, 1] + 1
keep = np.where((ws >= 25) & (hs >= 25))[0]
anchors = anchors[keep]
```
